Import libraries

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches

import numpy as np
import pandas as pd

from sklearn import linear_model
import statsmodels.formula.api as smf
```

Read data (numerical)

```python
df = pd.read_csv("../data/student/student-mat.csv", delimiter=";")
```

Linear Regression

```python
lm = smf.ols(formula='G3 ~ G1 + G2', data=df).fit()
lm.summary()
```

Plot

```python
X = np.array(df[['G1', 'G2']])
y = np.array(df['G3'])

regr = linear_model.LinearRegression()
regr.fit(X, y)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['G1'], df['G2'], df['G3'], c = 'blue')
ax.scatter(df['G1'], df['G2'], regr.predict(X), c = 'yellow')

blue_patch = mpatches.Patch(color='blue', label='Real Data')
yellow_patch = mpatches.Patch(color='yellow', label='Predicted Data')
plt.legend(handles=[blue_patch, yellow_patch])

plt.show()
```
