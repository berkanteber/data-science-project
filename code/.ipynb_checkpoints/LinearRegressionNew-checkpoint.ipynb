```python
from matplotlib import pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("../data/minimizeddata_full.csv", index_col = 0)
```

```python
lm_pisaread = smf.ols(formula='pisa_read ~ pryntry', data=df[['pisa_read', 'pryntry']]).fit()
lm_pisaread.summary()
```

```python
lm_pisamath = smf.ols(formula='pisa_math ~ pryntry', data=df[['pisa_math', 'pryntry']]).fit()
lm_pisamath.summary()
```

```python
lm_pisascience = smf.ols(formula='pisa_science ~ pryntry', data=df[['pisa_science', 'pryntry']]).fit()
lm_pisascience.summary()
```

    Regressions on pisa_read, pisa_math, pisa_science are all non-significant.

```python
lm_rdexp_unrestricted = smf.ols(formula='rd_exp ~ pisa_read + pisa_math + pisa_science + pryntry + try_public + try_private', data=df[['rd_exp', 'pisa_read', 'pisa_math', 'pisa_science', 'pryntry', 'try_public', 'try_private']]).fit()
lm_rdexp_unrestricted.summary()
```

```python
lm_rdexp_restrictallbutpisamathandtrypublic = smf.ols(formula='rd_exp ~ pisa_math + try_public', data=df[['rd_exp', 'pisa_math', 'try_public', 'try_private']]).fit()
lm_rdexp_unrestricted.compare_f_test(lm_rdexp_restrictallbutpisamathandtrypublic)
```

```python
lm_rdexp = lm_rdexp_restrictallbutpisamathandtrypublic
lm_rdexp.summary()
```

When I perform a f-test on unrestricted and restricted regressions on rd_exp, I see that all random variables except pisa_math, try_oublic and try_private are jointly non-significant.
Therefore, I took the restricted regression.

```python
lm_researchersinrd_unrestricted = smf.ols(formula='researchers_rd_permillionpeople ~ pisa_read + pisa_math + pisa_science + pryntry + try_public + try_private', data=df[['researchers_rd_permillionpeople', 'pisa_read', 'pisa_math', 'pisa_science', 'pryntry', 'try_public', 'try_private']]).fit()
lm_researchersinrd_unrestricted.summary()
```

```python
lm_researchersinrd_restrictallbuttrypublic = smf.ols(formula='rd_exp ~ try_public', data=df[['rd_exp','try_public']]).fit()
lm_researchersinrd_unrestricted.compare_f_test(lm_researchersinrd_restrictallbuttrypublic)
```
