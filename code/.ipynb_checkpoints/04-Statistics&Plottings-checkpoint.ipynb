Import libraries

```python
from matplotlib import pyplot as plt
from matplotlib import cm as cm
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
```

Read .csv file

```python
df = pd.read_csv("../data/alldata_full.csv", index_col = 0)
```

Convert data frame into a matrix and a list

```python
dm = df.values
header = df.axes[1].tolist()
data = dm.tolist()
```

Shorten header names

```python
short = header[:]
short[12] = "researchers_rd"
short[13] = "htech_exports"
short[14] = "paten_res"
short[15] = "patent_nres"
short[16] = "gdp_percapita"

df_short = pd.DataFrame(dm[:, 3:].astype(float), columns = short[3:], index = np.arange(1, 27))
```

Calculate correlation matrix

```python
correlation_df = df_short.corr()
correlation = correlation_df.values
```

Draw correlation matrix with coloring

```python
fig = plt.figure(figsize = (30, 30))
ax = fig.add_subplot(111)
cax = ax.matshow(correlation, vmin=-1, vmax=1, cmap = cm.seismic)
fig.colorbar(cax)
ticks = np.arange(0,15,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(short[3:])
ax.set_yticklabels(short[3:])

for i in range(15):
    for j in range(15):
        plt.annotate(round(correlation[i, j], 2), xy = (i, j), va = "center", ha = "center", color = "yellow", weight = "bold")
        
plt.show()
```

Draw scatter plot matrix (Diagonal: Histogram)

```python
scatter_matrix(df_short, figsize = (30, 30), diagonal = "hist")

plt.show()
```
