Import libraries

```python
from matplotlib import pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
from sklearn import linear_model
```

Read data

```python
df = pd.read_csv("../data/minimizeddata_full.csv", index_col = 0)
```

Regression: X = pryntry 
            Y = pisa_read, pisa_math, pisa_science 

```python
pryntry = np.array(df[['pryntry']])
pisa = np.array(df[['pisa_read', 'pisa_math', 'pisa_science']])

regr = linear_model.LinearRegression()
regr.fit(pryntry, pisa)

# The intercept
intercept = [round(x, 3) for x in regr.intercept_.tolist()]
print('Intercepts:', intercept)
# The coefficients
coefficient = [round(x, 3) for x in regr.coef_.tolist()[0]]
print('Coefficients:', coefficient)

# Explained variance score
print('R^2: %.3f' % regr.score(pryntry, pisa))

# Standard deviation
var = pisa.var()
stdev = var ** (1/2)
```

Comment: R^2 is only 0.12, it means that the spending on primary to nontertiary education explains only
         some part of the PISA Scores.

Plot

```python
# Plot outputs
labels = np.array(df[['name']]).flatten()

minindex = np.argmin(pryntry)

fig = plt.figure(figsize=(30, 30))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.set_title("PISA Read")
ax1.scatter(pryntry, pisa[:, 0],  color='black')

ax1.plot(pryntry, regr.predict(pryntry)[:, 0], color='blue', linewidth=2)
ax1.annotate(r'0 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 0]), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax1.plot(pryntry, regr.predict(pryntry)[:, 0] - stdev, color='lightblue', linewidth=1, linestyle='--')
ax1.annotate(r'-1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 0] - stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax1.plot(pryntry, regr.predict(pryntry)[:, 0] + stdev, color='lightblue', linewidth=1, linestyle='--')
ax1.annotate(r'+1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 0] + stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax1.plot(pryntry, regr.predict(pryntry)[:, 0] - 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax1.annotate(r'-2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 0] - 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax1.plot(pryntry, regr.predict(pryntry)[:, 0] + 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax1.annotate(r'+2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 0] + 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

for label, i,j in zip(labels, pryntry, pisa[:, 0]):
    ax1.annotate(label, xy=(i,j), xytext =(-3, -1), textcoords=('offset points'), ha='right', va='top')

ax1.set_xlabel("Spending on Primary to Nontertiary Education")    
ax1.set_ylabel("PISA Read Score")
    
ax2.set_title("PISA Math")
ax2.scatter(pryntry, pisa[:, 1],  color='black')

ax2.plot(pryntry, regr.predict(pryntry)[:, 1], color='blue', linewidth=2)
ax2.annotate(r'0 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 1]), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax2.plot(pryntry, regr.predict(pryntry)[:, 1] - stdev, color='lightblue', linewidth=1, linestyle='--')
ax2.annotate(r'-1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 1] - stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax2.plot(pryntry, regr.predict(pryntry)[:, 1] + stdev, color='lightblue', linewidth=1, linestyle='--')
ax2.annotate(r'+1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 1] + stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax2.plot(pryntry, regr.predict(pryntry)[:, 1] - 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax2.annotate(r'-2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 1] - 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax2.plot(pryntry, regr.predict(pryntry)[:, 1] + 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax2.annotate(r'+2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 1] + 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')


for label, i,j in zip(labels, pryntry, pisa[:, 1]):
    ax2.annotate(label, xy=(i,j), xytext =(-3, -1), textcoords=('offset points'), ha='right', va='top')

ax2.set_xlabel("Spending on Primary to Nontertiary Education")    
ax2.set_ylabel("PISA Math Score")    
    
ax3.set_title("PISA Science")
ax3.scatter(pryntry, pisa[:, 2],  color='black')

ax3.plot(pryntry, regr.predict(pryntry)[:, 2], color='blue', linewidth=2)
ax3.annotate(r'0 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 2]), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax3.plot(pryntry, regr.predict(pryntry)[:, 2] - stdev, color='lightblue', linewidth=1, linestyle='--')
ax3.annotate(r'-1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 2] - stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax3.plot(pryntry, regr.predict(pryntry)[:, 2] + stdev, color='lightblue', linewidth=1, linestyle='--')
ax3.annotate(r'+1 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 2] + stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax3.plot(pryntry, regr.predict(pryntry)[:, 2] - 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax3.annotate(r'-2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 2] - 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax3.plot(pryntry, regr.predict(pryntry)[:, 2] + 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax3.annotate(r'+2 $\sigma$', xy= (pryntry[minindex], regr.predict(pryntry)[minindex, 2] + 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

for label, i,j in zip(labels, pryntry, pisa[:, 2]):
    ax3.annotate(label, xy=(i,j), xytext =(-3, -1), textcoords=('offset points'), ha='right', va='top')

ax3.set_xlabel("Spending on Primary to Nontertiary Education")    
ax3.set_ylabel("PISA Science Score")    

plt.show()
```

Comment: As we can see in the figures, there are some outliers such as TUR and COL (2 out of 26),
         Also the there are still some variation except these countries.
                            

Regression: X = pryntry, try_public, try_private 
            Y = rd_exp

```python
edu_withnames = np.array(df[['name', 'pryntry', 'try_public', 'try_private']])
edu_sorted = edu_withnames[np.argsort(edu_withnames[:, 1:].sum(axis=1))]
edu = edu_sorted[:, 1:]
rd_exp = np.array(df[['rd_exp']])

regr = linear_model.LinearRegression()
regr.fit(edu, rd_exp)

# The intercept
intercept = [round(x, 2) for x in regr.intercept_.tolist()]
print('Intercepts:', intercept)
# The coefficients
coefficient = [round(x, 2) for x in regr.coef_.tolist()[0]]
print('Coefficients:', coefficient)

# Explained variance score
print('R^2: %.3f' % regr.score(edu, rd_exp))

# Standard deviation
var = rd_exp.var()
stdev = var ** (1/2)
```

Comment: R^2 is only 0.08, it means that the spending on primary to tertiary education is not enough
         to explain the variance in RD Expenditures.

Plot

```python
# Plot outputs
labels = edu_sorted[:, 0]
edu_total = [x[0] + x[1] + x[2] for x in edu]
edu_total = [float("%.1f" % x) for x in edu_total]

fig = plt.figure(figsize=(30, 10))
ax = fig.add_subplot(111)

ax.scatter(edu_total, rd_exp,  color='black')

ax.plot(edu_total, regr.predict(edu)[:, 0], color='blue', linewidth=2)
ax.annotate(r'0 $\sigma$', xy= (edu_total[0], regr.predict(edu)[0, 0]), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(edu_total, regr.predict(edu)[:, 0] - stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'-1 $\sigma$', xy= (edu_total[0], regr.predict(edu)[0, 0] - stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(edu_total, regr.predict(edu)[:, 0] + stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'+1 $\sigma$', xy= (edu_total[0], regr.predict(edu)[0, 0] + stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(edu_total, regr.predict(edu)[:, 0] - 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'-2 $\sigma$', xy= (edu_total[0], regr.predict(edu)[0, 0] - 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(edu_total, regr.predict(edu)[:, 0] + 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'+2 $\sigma$', xy= (edu_total[0], regr.predict(edu)[0, 0] + 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')


for label, i,j in zip(labels, edu_total, rd_exp):
    ax.annotate(label, xy=(i,j), xytext =(-3, -1), textcoords=('offset points'), ha='right', va='top')

ax.set_xlabel("Total Spending")    
ax.set_ylabel("RD Expenditures")

plt.show()
```

Comment: As we can see in the figures, although there are no outliers (2 stdev far from expected),
         there are still a lot variation that cannot explained by this model.

            Y = gini_index

```python
X_withnames = np.array(df[['name', 'pryntry', 'try_public', 'try_private']])
X_sorted = X_withnames[np.argsort(X_withnames[:, 1:].sum(axis=1))]
X = X_sorted[:, 1:]
Y = np.array(df[['gini_index']])

regr = linear_model.LinearRegression()
regr.fit(X, Y)

# The intercept
intercept = [round(x, 3) for x in regr.intercept_.tolist()]
print('Intercepts:', intercept)
# The coefficients
coefficient = [round(x, 3) for x in regr.coef_.tolist()[0]]
print('Coefficients:', coefficient)
# Standard Deviations


# Explained variance score
print('R^2: %.3f' % regr.score(X, Y))


# Standard deviation
var = Y.var()
stdev = var ** (1/2)
```

Comment: We can see that only education spending data explains a great portion of variation in GINI index.

Plot

```python
# Plot outputs
labels = X_sorted[:, 0]
X_total = [x[0] + x[1] + x[2] for x in X]
X_total = [float("%.1f" % x) for x in X_total]

fig = plt.figure(figsize=(30, 10))
ax = fig.add_subplot(111)

ax.scatter(X_total, Y,  color='black')

ax.plot(X_total, regr.predict(X)[:, 0], color='blue', linewidth=2)
ax.annotate(r'0 $\sigma$', xy= (X_total[0], regr.predict(X)[0, 0]), xytext =(-3, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(X_total, regr.predict(X)[:, 0] - stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'-1 $\sigma$', xy= (X_total[0], regr.predict(X)[0, 0] - stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(X_total, regr.predict(X)[:, 0] + stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'+1 $\sigma$', xy= (X_total[0], regr.predict(X)[0, 0] + stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(X_total, regr.predict(X)[:, 0] - 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'-2 $\sigma$', xy= (X_total[0], regr.predict(X)[0, 0] - 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

ax.plot(X_total, regr.predict(X)[:, 0] + 2 * stdev, color='lightblue', linewidth=1, linestyle='--')
ax.annotate(r'+2 $\sigma$', xy= (X_total[0], regr.predict(X)[0, 0] + 2 * stdev), xytext =(-10, 0), textcoords=('offset points'), ha='right', va='center')

for label, i,j in zip(labels, X_total, Y):
    ax.annotate(label, xy=(i,j), xytext =(-3, -1), textcoords=('offset points'), ha='right', va='top')

ax.set_xlabel("Total Spending")    
ax.set_ylabel("GINI Index")

plt.show()
```
